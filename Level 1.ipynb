```python
# Retrieve all orders with status "Completed"
cursor.execute("""
    SELECT orders.*
    FROM orders
    JOIN order_statuses ON orders.order_status = order_statuses.status_id
    WHERE order_statuses.status_description = 'Completed';
""")
completed_orders = cursor.fetchall()

# Print the results
for order in completed_orders:
    print(order)




```

```python
# Retrieve all orders with status "Rejected"
cursor.execute("""
    SELECT orders.*
    FROM orders
    JOIN order_statuses ON orders.order_status = order_statuses.status_id
    WHERE order_statuses.status_description = 'Rejected';
""")
rejected_orders = cursor.fetchall()

# Print the results
for order in rejected_orders:
    print(order)
```
