```python
import mysql.connector

# Set up the database connection
db_connection = mysql.connector.connect(
    host="localhost",
    user="root",
    password="",
    database="bikeshahd"
)

# Check if the connection is successful
if db_connection.is_connected():
    print("Connected to the database") 

# Create a cursor object to interact with the database
cursor = db_connection.cursor()

# Corrected query to retrieve customer names with their total orders and order IDs
query = """
    SELECT customers.first_name, orders.order_id, COUNT(orders.order_id) OVER (PARTITION BY orders.customer_id) AS total_orders
    FROM orders
    JOIN customers ON orders.customer_id = customers.customer_id
"""

# Execute the query
cursor.execute(query)

# Fetch all results
data = cursor.fetchall()

# Print the results
if data:
    for row in data:
        print(f"Customer Name: {row[0]}, Order ID: {row[1]}, Total Orders: {row[2]}")
else:
    print("No data available")

# Close the cursor and the connection
cursor.close()
db_connection.close()
```

```python
import mysql.connector

# Set up the database connection
db_connection = mysql.connector.connect(
    host="localhost",
    user="root",
    password="",
    database="bikeshahd"
)

# Check if the connection is successful
if db_connection.is_connected():
    print("Connected to the database") 

# Create a cursor object to interact with the database
cursor = db_connection.cursor()

# Query to retrieve product names with stock quantities less than 50
query = """
    SELECT products.product_name, stocks.quantity
    FROM products
    JOIN stocks ON products.product_id = stocks.product_id
    WHERE stocks.quantity < 50
"""

# Execute the query
cursor.execute(query)

# Fetch all results
data = cursor.fetchall()

# Print the results
if data:
    for row in data:
        print(f"Product Name: {row[0]}, Stock Quantity: {row[1]}")
else:
    print("No products with stock quantity less than 50")

# Close the cursor and the connection
cursor.close()
db_connection.close()
```
